```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Define uma semente aleatria para garantir a reprodutibilidade dos resultados
np.random.seed(42)
# Criao de dois conjuntos de dados fictcios, cada um com uma distribuio gaussiana

data1 = np.random.normal(loc=-5, scale=1, size=300)# 300 pontos com mdia -5 e desvio padro 1
data2 = np.random.normal(loc=5, scale=1, size=700) # 700 pontos com mdia 5 e desvio padro 1

# Combina os dois conjuntos de dados para formar um nico conjunto
data = np.concatenate([data1, data2])

# Visualiza o conjunto de dados original com um histograma
plt.hist(data, bins=30, density=True)
plt.title("Conjunto de Dados Original")
plt.show()

# Cria e treina um modelo de mistura gaussiana com dois componentes (dois picos)
gmm = GaussianMixture(n_components=2)
gmm.fit(data.reshape(-1, 1))  # Ajusta o modelo aos dados

# Usa o modelo treinado para gerar 1000 novos pontos de dados
samples = gmm.sample(1000)[0]

# Visualiza os dados gerados em comparao com os dados originais usando histogramas
plt.hist(samples, bins=30, density=True, alpha=0.5, color='blue', label='Dados Gerados')
plt.hist(data, bins=30, density=True, alpha=0.5, color='red', label='Dados Originais')
plt.title("Comparao: Dados Originais x Dados Gerados pelo GMM")
plt.legend()
plt.show()
```
